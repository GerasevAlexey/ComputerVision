```python
import os
import cv2
import stitching
import matplotlib.pyplot as plt
```

```python
curpath = os.getcwd()
orig = f'{curpath}//6//Panorama.jpg'
left = f'{curpath}//6//Left.jpg'
right = f'{curpath}//6//Right.jpg'

image = cv2.cvtColor(cv2.imread(orig), cv2.COLOR_BGR2RGB)
_, axes = plt.subplots(figsize=(20,20))
axes.imshow(image)
```

```python
sift_setup = {"detector": "sift", "confidence_threshold": 0.2}
orb_setup = {"detector": "orb", "confidence_threshold": 0.2}
orb_stitcher = stitching.Stitcher(**orb_setup)
sift_stitcher = stitching.Stitcher(**sift_setup)

result_sift = cv2.cvtColor(sift_stitcher.stitch([left, right]), cv2.COLOR_BGR2RGB)
_, axes = plt.subplots(figsize=(20,20))
axes.imshow(result_sift)
```

```python
result_orb = cv2.cvtColor(orb_stitcher.stitch([left, right]), cv2.COLOR_BGR2RGB)
_, axes = plt.subplots(figsize=(20,20))
axes.imshow(result_orb)
```
