```python
import os 
import cv2 
import matplotlib.pyplot as plt 
```

```python
cv2_dir = os.path.dirname(os.path.abspath(cv2.__file__))
print(cv2_dir)
haar_def = os.path.join(cv2_dir, 'data\haarcascade_frontalface_default.xml')
haar_eye = os.path.join(cv2_dir, 'data\haarcascade_eye.xml')
haar_smile = os.path.join(cv2_dir, 'data\haarcascade_smile.xml')
haar_frontalface1 = os.path.join(cv2_dir, 'data\haarcascade_frontalcatface.xml')
haar_upperbody = os.path.join(cv2_dir, 'data\haarcascade_upperbody.xml')
curpath = os.getcwd()
#print(curpath)
path = f'{curpath}//5//1.jpg'
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
face_cascade = cv2.CascadeClassifier(haar_def)
faces_rect = face_cascade.detectMultiScale(cv2.cvtColor(img, cv2.COLOR_RGB2GRAY), scaleFactor=1.1, minNeighbors=9)
res_img = img.copy()
for (x, y, w, h) in faces_rect:
    cv2.rectangle(res_img, (x, y), (x+w, y+h), (0, 255, 0), 2)
plt.imshow(res_img)
plt.show()
```

```python
cascade = cv2.CascadeClassifier(haar_smile)
rect = cascade.detectMultiScale(cv2.cvtColor(img, cv2.COLOR_RGB2GRAY), scaleFactor=1.8, minNeighbors=1)
res_img = img.copy()
for (x, y, w, h) in rect:
    cv2.rectangle(res_img, (x, y), (x+w, y+h), (0, 255, 0), 2)
plt.imshow(res_img)
plt.show()
```

```python
cascade = cv2.CascadeClassifier(haar_eye)
rect = cascade.detectMultiScale(cv2.cvtColor(img, cv2.COLOR_RGB2GRAY), scaleFactor=1.8, minNeighbors=1, maxSize = (10, 10))
res_img = img.copy()
for (x, y, w, h) in rect:
    cv2.rectangle(res_img, (x, y), (x+w, y+h), (0, 255, 0), 2)
plt.imshow(res_img)
plt.show()
```

```python
cascade = cv2.CascadeClassifier(haar_upperbody)
rect = cascade.detectMultiScale(cv2.cvtColor(img, cv2.COLOR_RGB2GRAY), scaleFactor=1.8, minNeighbors=1, maxSize = (10, 10))
res_img = img.copy()
for (x, y, w, h) in rect:
    cv2.rectangle(res_img, (x, y), (x+w, y+h), (0, 255, 0), 2)
plt.imshow(res_img)
plt.show()
```
